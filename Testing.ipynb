```python
import os
import sys
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
import librosa
import librosa.core
import librosa.feature
import numpy
```

```python
path = os.path.abspath("00000000.wav")

wavfile = read(path)
```

```python
wavfile.shape()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys


spf = wave.open("wavfile.wav", "r")

# Extract Raw Audio from Wav File
signal = spf.readframes(-1)
signal = np.fromstring(signal, "Int16")


# If Stereo
if spf.getnchannels() == 2:
    print("Just mono files")
    sys.exit(0)

plt.figure(1)
plt.title("Signal Wave...")
plt.plot(signal)
plt.show()
```

```python

plt.plot(wavfile)
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```

```python
#loading the wav file

wav_name = librosa.load("abnormal.wav", sr=None, mono=False)
wav_name_plus6 = librosa.load("abnormal6.wav", sr=None, mono=False)
wav_name_min6 = librosa.load("abnormalmin6.wav", sr=None, mono=False)

#demultiplexing the wav file

channel=0

multi_channel_data, sr = wav_name
if multi_channel_data.ndim <= 1:
    sr, multi_channel_data

sr,y = sr, numpy.array(multi_channel_data)[channel, :]

#creating a melspectogram

mel_spectrogram = librosa.feature.melspectrogram(y=y,
                                                     sr=sr,
                                                     n_fft=1024,
                                                     hop_length=512,
                                                     n_mels=64,
                                                     power=2.0)

import matplotlib.pyplot as plt
from librosa import display
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(mel_spectrogram, ref=numpy.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='abnormal.wav')

multi_channel_data, sr = wav_name_plus6
if multi_channel_data.ndim <= 1:
    sr, multi_channel_data

sr,y = sr, numpy.array(multi_channel_data)[channel, :]

#creating a melspectogram

mel_spectrogram_plus6 = librosa.feature.melspectrogram(y=y,
                                                     sr=sr,
                                                     n_fft=1024,
                                                     hop_length=512,
                                                     n_mels=64,
                                                     power=2.0)

import matplotlib.pyplot as plt
from librosa import display
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(mel_spectrogram_plus6, ref=numpy.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='abnormalmin6.wav')

multi_channel_data, sr = wav_name_min6
if multi_channel_data.ndim <= 1:
    sr, multi_channel_data

sr,y = sr, numpy.array(multi_channel_data)[channel, :]

#creating a melspectogram

mel_spectrogram_min6 = librosa.feature.melspectrogram(y=y,
                                                     sr=sr,
                                                     n_fft=1024,
                                                     hop_length=512,
                                                     n_mels=64,
                                                     power=2.0)
import matplotlib.pyplot as plt
from librosa import display
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(mel_spectrogram_min6, ref=numpy.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='abnormalplus6.wav')
```
